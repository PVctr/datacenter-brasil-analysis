```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

sns.set_theme(style="whitegrid", palette="muted")
plt.rcParams['figure.figsize'] = (12, 5)
plt.rcParams['font.size'] = 12

print("Bibliotecas carregadas com sucesso!")
```

```python
# Dataset construído com dados reais de fontes oficiais
# Fontes: Brasscom, Mordor Intelligence, AbraCloud, Moody's Local (2024/2025)

# Evolução do mercado de data centers no Brasil (receita em US$ bilhões)
mercado_dc = pd.DataFrame({
    'Ano': [2020, 2021, 2022, 2023, 2024, 2025, 2026, 2027, 2028, 2029],
    'Receita_USD_bi': [6.35, 7.10, 7.95, 8.80, 10.86, 12.10, 14.50, 17.20, 20.80, 28.46],
    'Capacidade_MW':  [400,  460,  520,  610,  740,   850,   950,  1050,  1150,  1210],
    'Tipo': ['Real','Real','Real','Real','Real','Projeção','Projeção','Projeção','Projeção','Projeção']
})

# Investimentos anunciados por empresa (em R$ bilhões)
investimentos = pd.DataFrame({
    'Empresa':     ['Microsoft', 'Amazon (AWS)', 'Scala', 'Ascenty', 'Grupo ClickIP', 'BNDES (crédito)'],
    'Valor_R_bi':  [14.7,        10.1,           3.0,     0.3,        0.02,            2.0],
    'Ano_anuncio': [2024,        2024,           2024,    2025,       2024,            2024],
    'Foco':        ['Cloud + IA', 'Cloud + IA', 'IA + Hyperscale', 'Cloud', 'Edge/Norte', 'Fomento']
})

# Faturamento cloud nacional (R$ bilhões) — fonte: AbraCloud
cloud_nacional = pd.DataFrame({
    'Ano':           [2022, 2023, 2024, 2025],
    'Faturamento_bi': [1.4,  1.8,  2.1,  3.0],
    'Status':        ['Real','Real','Real','Projeção']
})

# Participação regional na capacidade instalada
regioes = pd.DataFrame({
    'Regiao':    ['São Paulo', 'Rio de Janeiro', 'Sul + MG + CE', 'Outros'],
    'Perc_2024': [80, 7, 8, 5],
    'Perc_2028': [80, 10, 10, 0]  # projeção Oliver Wyman
})

print("Dataset montado com sucesso!")
print(f"\nMercado DC: {len(mercado_dc)} anos de dados (2020–2029)")
print(f"Investimentos mapeados: {len(investimentos)} empresas")
print(f"Cloud nacional: {len(cloud_nacional)} anos de dados")
```

```python
# Separando dados reais das projeções para visualizar diferente
reais = mercado_dc[mercado_dc['Tipo'] == 'Real']
projecoes = mercado_dc[mercado_dc['Tipo'] == 'Projeção']

fig, ax1 = plt.subplots()

# Barras para dados reais (roxo) e projeções (mais claro)
ax1.bar(reais['Ano'], reais['Receita_USD_bi'], color='#7F77DD', label='Receita real')
ax1.bar(projecoes['Ano'], projecoes['Receita_USD_bi'], color='#CECBF6', label='Projeção')

# Segunda linha com capacidade em MW (eixo direito)
ax2 = ax1.twinx()
ax2.plot(mercado_dc['Ano'], mercado_dc['Capacidade_MW'],
         color='#1D9E75', marker='o', linewidth=2, label='Capacidade (MW)')

# Linha separando real de projeção
ax1.axvline(x=2024.5, color='gray', linestyle='--', linewidth=1, alpha=0.6)
ax1.text(2024.6, 25, 'Projeção →', fontsize=10, color='gray')

ax1.set_title('Mercado de data centers no Brasil (2020–2029)', fontsize=14, pad=12)
ax1.set_xlabel('Ano')
ax1.set_ylabel('Receita (US$ bilhões)')
ax2.set_ylabel('Capacidade instalada (MW)', color='#1D9E75')
ax1.legend(loc='upper left')
ax2.legend(loc='center left')
plt.tight_layout()
plt.savefig('mercado_dc.png', dpi=150)
plt.show()

print(f"Crescimento 2020–2024: {((10.86/6.35)-1)*100:.1f}%")
print(f"Crescimento projetado 2024–2029: {((28.46/10.86)-1)*100:.1f}%")
```

```python
cores_status = ['#7F77DD' if s == 'Real' else '#CECBF6' for s in cloud_nacional['Status']]

fig, ax = plt.subplots()
bars = ax.bar(cloud_nacional['Ano'], cloud_nacional['Faturamento_bi'], color=cores_status)

# Valor em cima de cada barra
for bar, val in zip(bars, cloud_nacional['Faturamento_bi']):
    ax.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 0.05,
            f'R$ {val:.1f} bi', ha='center', fontsize=11)

ax.axvline(x=2024.5, color='gray', linestyle='--', linewidth=1, alpha=0.6)
ax.text(2024.6, 2.7, 'Projeção →', fontsize=10, color='gray')

ax.set_title('Faturamento do cloud nacional brasileiro (2022–2025)', fontsize=14, pad=12)
ax.set_xlabel('Ano')
ax.set_ylabel('Faturamento (R$ bilhões)')
ax.set_ylim(0, 3.8)
plt.tight_layout()
plt.savefig('cloud_nacional.png', dpi=150)
plt.show()

crescimento = ((2.1 - 1.4) / 1.4) * 100
print(f"Crescimento 2022–2024: {crescimento:.1f}%")
print(f"Projeção 2025: R$ 3,0 bilhões (+{((3.0-2.1)/2.1)*100:.1f}% vs 2024)")
```
